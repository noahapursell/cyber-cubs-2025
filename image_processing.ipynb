```python
from lib.image_transforms.add_noise import add_pixel_noise
from lib.image_transforms.remove_blocks import split_and_remove
import os

IMAGES_DIR = "base_images"
REMOVED_DIR = "removed_images"
NOISE_DIR = "noised_images"

for dir_name in [IMAGES_DIR, REMOVED_DIR, NOISE_DIR]:
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
```

```python
import os

for file in os.listdir(IMAGES_DIR):
    # for prob_level in [1, 0.9, 0.8, 0.5, 0.25, 0]:
    for prob_level in [1, 0.95, 0.9, 0.75, 0]:
        # Create directories if they don't exist
        removed_subdir = os.path.join(REMOVED_DIR, f"{file[:-4]}")
        noise_subdir = os.path.join(NOISE_DIR, f"{file[:-4]}")
        
        os.makedirs(removed_subdir, exist_ok=True)
        os.makedirs(noise_subdir, exist_ok=True)

        # Process and save images
        split_and_remove(
            image_path=os.path.join(IMAGES_DIR, file),
            probability=prob_level,
            grid_size=15,
            output_path=os.path.join(removed_subdir, f"at{prob_level}.png")
        )

        add_pixel_noise(
            image_path=os.path.join(IMAGES_DIR, file),
            probability=prob_level,
            output_path=os.path.join(noise_subdir, f"at{prob_level}.png")
        )
```
